# Gimbal Control System Analysis
Import required libraries and load data from CSV file

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Read data from CSV
df = pd.read_csv('gimbal_response.csv')

# Calculate total error
df['err_total'] = np.sqrt(df['err_x_px']**2 + df['err_y_px']**2)

# Calculate accumulated errors
df['accumulate_err_x'] = df['err_x_px'].cumsum()
df['accumulate_err_y'] = df['err_y_px'].cumsum()
df['accumulate_err_total'] = np.sqrt(df['accumulate_err_x']**2 + df['accumulate_err_y']**2)

# Set plot style
plt.style.use('seaborn')
plt.rcParams['figure.figsize'] = [12, 6]
```

## 1. Tracking Errors
Plot instantaneous tracking errors in X and Y directions, and total error magnitude

```python
plt.figure(figsize=(12, 6))
plt.plot(df['t'], df['err_x_px'], label='Error X')
plt.plot(df['t'], df['err_y_px'], label='Error Y')
plt.plot(df['t'], df['err_total'], label='Total Error')
plt.xlabel('Time (s)')
plt.ylabel('Error (pixels)')
plt.title('Tracking Errors over Time')
plt.grid(True)
plt.legend()
plt.show()
```

## 2. Accumulated Error X
Plot accumulated error in X direction

```python
plt.figure(figsize=(12, 6))
plt.plot(df['t'], df['accumulate_err_x'], label='Accumulated Error X')
plt.xlabel('Time (s)')
plt.ylabel('Accumulated Error (pixels)')
plt.title('Accumulated Error in X Direction')
plt.grid(True)
plt.legend()
plt.show()
```

## 3. Accumulated Error Y
Plot accumulated error in Y direction

```python
plt.figure(figsize=(12, 6))
plt.plot(df['t'], df['accumulate_err_y'], label='Accumulated Error Y')
plt.xlabel('Time (s)')
plt.ylabel('Accumulated Error (pixels)')
plt.title('Accumulated Error in Y Direction')
plt.grid(True)
plt.legend()
plt.show()
```

## 4. Total Accumulated Error
Plot total accumulated error magnitude

```python
plt.figure(figsize=(12, 6))
plt.plot(df['t'], df['accumulate_err_total'], label='Total Accumulated Error')
plt.xlabel('Time (s)')
plt.ylabel('Accumulated Error Magnitude (pixels)')
plt.title('Total Accumulated Error over Time')
plt.grid(True)
plt.legend()
plt.show()
```

## 5. Controller Output X (Yaw)
Plot controller output signal for Yaw axis

```python
plt.figure(figsize=(12, 6))
plt.plot(df['t'], df['u_yaw_dps'], label='Controller Output X (Yaw)')
plt.xlabel('Time (s)')
plt.ylabel('Control Signal (deg/s)')
plt.title('Controller Output X (Yaw) over Time')
plt.grid(True)
plt.legend()
plt.show()
```

## 6. Controller Output Y (Pitch)
Plot controller output signal for Pitch axis

```python
plt.figure(figsize=(12, 6))
plt.plot(df['t'], df['u_pitch_dps'], label='Controller Output Y (Pitch)')
plt.xlabel('Time (s)')
plt.ylabel('Control Signal (deg/s)')
plt.title('Controller Output Y (Pitch) over Time')
plt.grid(True)
plt.legend()
plt.show()
```

Plot gimbal pitch angle over time

```python
plt.figure(figsize=(12, 6))
plt.plot(df['t'], df['pitch_angle'], label='Pitch Angle')
plt.xlabel('Time (s)')
plt.ylabel('Angle (deg)')
plt.title('Gimbal Pitch Angle over Time')
plt.grid(True)
plt.legend()
plt.show()
```

Plot gimbal yaw angle over time
